# Amazon Recommendation System using Python

```python
import numpy as np
import pandas as pd
df= pd.read_csv("ratings_Electronics (1).csv")
```

```python
df
```

```python
df.columns = ['user_id', 'product_id','ratings','timestamp']
```

```python
df
```

```python
df.isnull().sum()
```

```python
product_grouped =df.groupby('product_id', as_index=False).agg({'ratings':[np.size, np.sum, np.mean]})
```

```python
product_grouped
```

```python
popular_product=product_grouped.sort_values(('ratings','mean'),ascending=False)
```

```python
popular_product
```

```python
product_sum=product_grouped['ratings']['sum'].sum()
```

```python
product_sum
```

```python
popular_product['persentage']=product_grouped['ratings']['sum'].div(product_sum)*100
```

```python
popular_product
```

```python
popular_product=popular_product.sort_values(('persentage'),ascending=False)
```

```python
popular_product['Rank']=popular_product['persentage'].rank(ascending=False)
```
